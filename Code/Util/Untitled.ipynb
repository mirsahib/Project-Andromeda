```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(100,size=(1,5))
x
```

```python
plt.plot(x.T)
```

```python
npSin = np.sin(x)
plt.plot(npSin.T)
```

```python
sp = np.fft.fft(npSin)
```

```python
freq = np.fft.fftfreq(x.shape[-1])
```

```python
plt.plot(freq, sp.real, freq, sp.imag)
```
